```python
import pandas as pd

url = "https://kworb.net/spotify/songs_2025.html"
# read_html returns a list of all tables found on the page
tables = pd.read_html(url)

# The chart is the first table on this specific page
df = tables[0]

# Clean up column names and export
df.to_csv("top_spotify_songs_2025.csv", index=False)
print("CSV successfully created!")
```
